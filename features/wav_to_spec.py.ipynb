```python
def spectrogram(input_sample,file_path,path_save):
    name_file = file_path.split('/')[-1].split('.')[0] 
    freqs, time, spectrogram = signal.spectrogram(input_sample)
    spec = np.log(spectrogram.T.astype('float32')) #converting into logarithmic scale for better visualization
    plt.imshow(spec.T, extent = [time.min(), time.max(), freqs.min(), freqs.max()], origin = 'lower', aspect = 'auto')
    plt.savefig(path_save+'/'+name_file)
```
